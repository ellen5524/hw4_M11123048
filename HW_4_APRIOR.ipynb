import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from sklearn.model_selection import train_test_split

```python
dataset = pd.read_csv('C:/Users/liping/OneDrive/文件/學校資料/碩一上/資料探勘/Hw/dataminingHW4/交易資料集.csv', na_values = '?' )
```

```python
df = dataset
df = df.replace('?', np.NaN)
df = df.dropna()
df
```

```python
#刪除數量為零或負值的交易
df2 = df[df["QUANTITY"] > 0]
df2
```

```python
#相同INVOICE_NO表示同一筆交易紀錄，QUANTITY做加總
basket = (df2.groupby(['INVOICE_NO','ITEM_ID'])
          ['QUANTITY'].sum().unstack().reset_index().fillna(0).set_index('INVOICE_NO'))
basket
```

```python
#轉成布林型態
def encode_units(x):
    if x <= 0:
        return False
    if x >= 1:
        return True

basket_sets = basket.applymap(encode_units)
basket_sets
```

# APRIORI

```python
import datetime
start = datetime.datetime.now()
result = apriori(basket_sets, min_support=0.001, use_colnames=True)
rules = association_rules(result, metric="lift", min_threshold=1)
#結束時間
end = datetime.datetime.now()
#輸出程式花費時間
print("執行時間：", end - start)
rules
```

```python
rules.info()
```

```python
#儲存
newdf=rules
newdf=np.array(newdf)
np.save('apriori_arules001.npy',newdf)
```

```python
#讀取
rulest=pd.DataFrame(np.load('apriori_rules001.npy',allow_pickle=True))
rulest.columns = ['antecedents','consequents','antecedent support','consequent support','support','confidence','lift','leverage','conviction']
rulest
```

```python
#冗餘規則
for a in range(len(rulest['antecedents'])):
    for b in range(a+1,len(rulest['antecedents'])):
        if rulest['antecedents'][a] <= rulest['antecedents'][b] and rulest['consequents'][a] == rulest['consequents'][b] and rulest['confidence'][a] <= rulest['confidence'][b]:
            rulest.drop(index = a ,axis = 0, inplace=True)
            break
```
