### 同一交易紀錄

```python
# 建立list
record = []

# 放入list
for inv in df2["INVOICE_NO"].unique():
  record.append(df2[df2["INVOICE_NO"] == inv]["ITEM_ID"].values.tolist())
```

```python
record
```

### FP-Growth

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from datetime import datetime
```

```python
#進行 one-hot 編碼
te = TransactionEncoder()
te_ary = te.fit(record).transform(record)
data = pd.DataFrame(te_ary, columns=te.columns_)
data
```

```python
#起始時間
start = datetime.now()

#支持度
Support_itemsets = fpgrowth(data, min_support = 0.0008, use_colnames = True)

#關聯規則
Association_Rules = association_rules(Support_itemsets, metric = "lift", min_threshold = 1)

#結束時間
end = datetime.now()
time = end - start
print("執行時間：", time)

Association_Rules
```

```python
Association_Rules.to_excel("Association_Rules_FP-Growth.xlsx")
```

```python
#冗餘規則
for a in range(len(Association_Rules['antecedents'])):
    for b in range(a+1,len(Association_Rules['antecedents'])):
        if Association_Rules['antecedents'][a] <= Association_Rules['antecedents'][b] and Association_Rules['consequents'][a] == Association_Rules['consequents'][b] and Association_Rules['confidence'][a] <= Association_Rules['confidence'][b]:
            Association_Rules.drop(index = a ,axis = 0, inplace=True)
            break
```

```python
user_input = set(input("請輸入商品編號(不同商品以空白間隔開)").split(' '))
search_input = set(map(int, user_input))
print("您輸入了:",search_input)

rule = pd.DataFrame()
rule = Association_Rules[(Association_Rules['antecedents'] == search_input)]

kk = list(rule['consequents'])

if rule.empty:
    print("沒有推薦的商品")
else:
    print("推薦商品編號為：")
    for i in range(len(kk)):
        cons = kk[i]
        if len(cons) < 2:
            print(i+1,cons)
```
